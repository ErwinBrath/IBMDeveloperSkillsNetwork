```python
import pandas as pd
import numpy as np
from ipywidgets import interact, interactive, fixed, interact_manual
import matplotlib.pyplot as plt
import seaborn as sns

# Import clean data 
path = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/module_5_auto.csv'
df = pd.read_csv(path)
df=df._get_numeric_data()
df.head()
```

```python
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width = 15
    height = 5
    plt.figure(figsize=(width, height)) 

    ax1 = sns.distplot(RedFunction, hist=True, color="r", label=RedName, bins=100)
    ax2 = sns.distplot(BlueFunction, hist=True, color="b", label=BlueName, ax=ax1)

    plt.title(Title)

    plt.show()
    plt.close()
```

```python
RedFunction = df['price']
BlueFunction = []
RedName='namered'
BlueName='nameblue'
Title='hola amigos'
DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title)
```
