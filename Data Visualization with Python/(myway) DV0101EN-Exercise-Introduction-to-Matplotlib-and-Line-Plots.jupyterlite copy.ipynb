## Visualizar gráfico de lineas para haiti

```python
df_can.head()
```

```python
df = df_can.drop(['AREA','REG','Type','Coverage'], axis=1)
df.head()
```

```python
df = df.melt(id_vars=['OdName'	,'AreaName'	,'RegName'	,'DEV'	,'DevName'	], var_name='Year',value_name='Inmigrants')
df.head(3)
```

```python
df.dtypes
```

```python
dtypes = {
'OdName':        'string',
'AreaName':      'string',
'RegName':       'string',
'DEV':            'int64',
'DevName':       'string',
'Year':          'int64',
'Inmigrants':     'int64',
}
df = df.astype(dtypes)
df.head()
```

```python
df_haiti = df[df['OdName']=='Haiti']
df_haiti.head(10)
```

```python
df_haiti.plot(x='Year', y='Inmigrants')
```

```python
df_CI=df[df['OdName'].isin(['China','India'])]
df_CI.head()
```

```python
df_CI[['OdName','Year','Inmigrants']].pivot(columns=['OdName'], index='Year', values='Inmigrants').plot()
```
